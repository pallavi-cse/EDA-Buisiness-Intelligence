```python
import pandas as pd
df=pd.read_csv('/content/train.csv')
df.head()
```

```python
df.columns.tolist()
```

```python
import matplotlib.pyplot as plt
```

```python
print("Total Sales:", df['Sales'].sum())
print("Total Orders:", df['Order ID'].nunique())
print("Total Customers:", df['Customer ID'].nunique())
```

```python
# Sales by Region
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Region')
plt.show()

# Sales by Segment
df.groupby('Segment')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Segment')
plt.show()

# Top 10 States by Sales
df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 States by Sales')
plt.show()
```

## Key Insights

1. The dataset contains sales records from multiple regions and product categories.

2. Sales are not evenly distributed, with some orders generating significantly higher sales than others.

3. Different categories contribute differently to total sales.

4. Sales performance varies across regions, indicating differences in market demand.
